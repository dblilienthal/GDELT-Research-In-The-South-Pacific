This is all the Themes in the V2Themes

```python
result = {}

V2_Themes = df['V2Themes'].convert_dtypes()
V2themes_list = []

#print(type(temp_df))

V2_Themes.dropna(inplace=True)

for row in V2_Themes:
    # Split each row by the semicolon
    temp = row.split(";")

    # Add each theme to a result dictionary
    for themes in temp:
        theme = themes.split(',')
        result[theme[0]] = result.get(theme[0], 0) + 1

# Sort the results
sort_orders = sorted(result.items(), key=lambda x: x[1], reverse=True)

# Print out the top 50 results
for i in sort_orders:
    print(i[0], i[1])
    t_list = [i[0],i[1]]
    V2themes_list.append(t_list)
```

This is all the themes in Themes

```python
result = {}

Themes = df['Themes'].convert_dtypes()
Themes_list = []

#print(type(temp_df))

Themes.dropna(inplace=True)

for row in Themes:
    # Split each row by the semicolon
    temp = row.split(";")

    # Add each theme to a result dictionary
    for themes in temp:
        theme = themes.split(',')
        result[theme[0]] = result.get(theme[0], 0) + 1

# Sort the results
sort_orders = sorted(result.items(), key=lambda x: x[1], reverse=True)

# Print out the top 50 results
for i in sort_orders:
    print(i[0], i[1])
    t_list = [i[0],i[1]]
    Themes_list.append(t_list)
```

Combine both results and compare them side by side

```python
df_V2themes = pd.DataFrame(data=V2themes_list, columns=['Themes', 'Count'] )
df_themes = pd.DataFrame(data=Themes_list, columns=['Themes', 'Count'])
combined_themes = df_V2themes.merge(df_themes, on='Themes',suffixes=('_V2Themes', '_Themes'))
combined_themes['Difference'] = combined_themes['Count_V2Themes'] - combined_themes['Count_Themes']
combined_themes
```

Now i will try filtering the V2Themes to only allow one unique theme per row to be added. I wil then compare the difference again and see the difference

```python
result = {}

V2_Themes = df['V2Themes'].convert_dtypes()
V2themes_list_no_dup = []

unique_list = []

#print(type(temp_df))

V2_Themes.dropna(inplace=True)

for row in V2_Themes:
    # Split each row by the semicolon
    temp = row.split(";")

    temp_list = []
    # Add each theme to a result dictionary
    for themes in temp:
        # Split the themes by the comma
        theme = themes.split(',')
        # Add the themes to a temp list
        temp_list.append(theme[0])
        # Convert the list to a set to eliminate duplicates
        list_set = set(temp_list) 
        # Convert the set back to a list
        unique_list = (list(list_set)) 
    # Add each theme into a dictionary 
    for x in unique_list: 
        result[x] = result.get(x, 0) + 1

# Sort the results
sort_orders = sorted(result.items(), key=lambda x: x[1], reverse=True)

# Print out the top 50 results
for i in sort_orders:
    print(i[0], i[1])
    t_list = [i[0],i[1]]
    V2themes_list_no_dup.append(t_list)
```

```python
df_V2themes_no_dup = pd.DataFrame(data=V2themes_list_no_dup, columns=['Themes', 'Count'] )
df_themes = pd.DataFrame(data=Themes_list, columns=['Themes', 'Count'])
combined_themes = df_V2themes_no_dup.merge(df_themes, on='Themes',suffixes=('_V2Themes_No_Duplicates', '_Themes'))
combined_themes['Difference'] = combined_themes['Count_V2Themes_No_Duplicates'] - combined_themes['Count_Themes']
combined_themes
```
