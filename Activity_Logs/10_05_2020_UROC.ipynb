```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sqlalchemy
from sqlalchemy import create_engine
pd.set_option('display.max_rows', None)
```

### Load the Data 

```python
events = pd.read_csv("C:/Users/dlili/OneDrive/Documents/CSUMB/UROC Fall 2020/CSV Files/Events_Table_10042020.csv")
eventmentions = pd.read_csv("C:/Users/dlili/OneDrive/Documents/CSUMB/UROC Fall 2020/CSV Files/Eventmentions_Table_10042020.csv")
events_and_eventmentions = pd.merge(events, eventmentions, how='left', on='GLOBALEVENTID')

# Create the SQL engine 
engine = create_engine('sqlite://', echo=False)
events_and_eventmentions.to_sql('events_and_eventmentions', con=engine)
```

## Break the table into countries and territories

#### Get the countries

```python
query = """
SELECT GLOBALEVENTID, SQLDATE, MonthYear, Actor1Name, Actor1CountryCode, Actor2Name, Actor2CountryCode, EventCode, GoldsteinScale, NumMentions,
NumSources, NumArticles, AvgTone, Actor1Geo_FullName, Actor1Geo_CountryCode, Actor2Geo_FullName, Actor2Geo_CountryCode,  ActionGeo_FullName, ActionGeo_ADM1Code,
ActionGeo_ADM2Code, SOURCEURL, MentionTimeDate, MentionType, MentionSourceName, MentionIdentifier, Confidence, MentionDocLen, MentionDocTone
FROM events_and_eventmentions
WHERE ActionGeo_ADM1Code like 'PP%' OR 
    ActionGeo_ADM1Code like 'TN%' OR
    ActionGeo_ADM1Code like 'NH%' OR
    ActionGeo_ADM1Code like 'NE%' OR
    ActionGeo_ADM1Code like 'FP%' OR
    ActionGeo_ADM1Code like 'PS%' OR
    ActionGeo_ADM1Code like 'FJ%' OR
    ActionGeo_ADM1Code like 'NR%' OR
    ActionGeo_ADM1Code like 'ID%' OR
    ActionGeo_ADM1Code like 'FM%' OR
    ActionGeo_ADM1Code like 'TV%' OR
    ActionGeo_ADM1Code like 'BP%' OR
    ActionGeo_ADM1Code like 'CW%' OR
    ActionGeo_ADM1Code like 'KR%' OR
    ActionGeo_ADM1Code like 'RM%' OR
    ActionGeo_ADM1Code like 'WS%' 
"""

# Execute the query and save the result
temp_result = engine.execute(query).fetchall()
# Convert result into numpy array
temp1 = np.array(temp_result)
# Convert np array into dataframe
countries = pd.DataFrame(data=temp1, columns=['GLOBALEVENTID', 'SQLDATE', 'MonthYear', 'Actor1Name', 'Actor1CountryCode', 'Actor2Name', 'Actor2CountryCode', 'EventCode', 'GoldsteinScale', 'NumMentions' ,'NumSources', 'NumArticles', 'AvgTone', 'Actor1Geo_FullName', 'Actor1Geo_CountryCode', 'Actor2Geo_FullName', 'Actor2Geo_CountryCode',  'ActionGeo_FullName', 'ActionGeo_ADM1Code', 'ActionGeo_ADM2Code', 'SOURCEURL', 'MentionTimeDate', 'MentionType', 'MentionSourceName', 'MentionIdentifier', 'Confidence', 'MentionDocLen', 'MentionDocTone'])
# Convert the datatypes from Object 
countries = countries.convert_dtypes()
# Show the top 10 results
countries.head(10)
```
