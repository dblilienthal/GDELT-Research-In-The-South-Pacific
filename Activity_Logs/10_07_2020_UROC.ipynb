### Split the events and eventmentions table into countries and territories. 

```python
query = """
SELECT GLOBALEVENTID, SQLDATE, MonthYear, Actor1Name, Actor1CountryCode, Actor2Name, Actor2CountryCode, EventCode, GoldsteinScale, NumMentions,
NumSources, NumArticles, AvgTone, Actor1Geo_FullName, Actor1Geo_CountryCode, Actor2Geo_FullName, Actor2Geo_CountryCode,  ActionGeo_FullName, ActionGeo_ADM1Code,
ActionGeo_ADM2Code, SOURCEURL, MentionTimeDate, MentionType, MentionSourceName, MentionIdentifier, Confidence, MentionDocLen, MentionDocTone
FROM events_and_eventmentions
WHERE ActionGeo_ADM1Code like 'PP%' OR 
    ActionGeo_ADM1Code like 'TN%' OR
    ActionGeo_ADM1Code like 'NH%' OR
    ActionGeo_ADM1Code like 'NE%' OR
    ActionGeo_ADM1Code like 'FP%' OR
    ActionGeo_ADM1Code like 'PS%' OR
    ActionGeo_ADM1Code like 'FJ%' OR
    ActionGeo_ADM1Code like 'NR%' OR
    ActionGeo_ADM1Code like 'ID%' OR
    ActionGeo_ADM1Code like 'FM%' OR
    ActionGeo_ADM1Code like 'TV%' OR
    ActionGeo_ADM1Code like 'BP%' OR
    ActionGeo_ADM1Code like 'CW%' OR
    ActionGeo_ADM1Code like 'KR%' OR
    ActionGeo_ADM1Code like 'RM%' OR
    ActionGeo_ADM1Code like 'WS%' 
"""

# Execute the query and save the result
temp_result = engine.execute(query).fetchall()
# Convert result into numpy array
temp1 = np.array(temp_result)
# Convert np array into dataframe
countries = pd.DataFrame(data=temp1, columns=['GLOBALEVENTID', 'SQLDATE', 'MonthYear', 'Actor1Name', 'Actor1CountryCode', 'Actor2Name', 'Actor2CountryCode', 'EventCode', 'GoldsteinScale', 'NumMentions' ,'NumSources', 'NumArticles', 'AvgTone', 'Actor1Geo_FullName', 'Actor1Geo_CountryCode', 'Actor2Geo_FullName', 'Actor2Geo_CountryCode',  'ActionGeo_FullName', 'ActionGeo_ADM1Code', 'ActionGeo_ADM2Code', 'SOURCEURL', 'MentionTimeDate', 'MentionType', 'MentionSourceName', 'MentionIdentifier', 'Confidence', 'MentionDocLen', 'MentionDocTone'])
# Convert the datatypes from Object 
countries = countries.convert_dtypes()
# Show the top 10 results
countries.head(10)
```

```python
query = """
SELECT GLOBALEVENTID, SQLDATE, MonthYear, Actor1Name, Actor1CountryCode, Actor2Name, Actor2CountryCode, EventCode, GoldsteinScale, NumMentions,
NumSources, NumArticles, AvgTone, Actor1Geo_FullName, Actor1Geo_CountryCode, Actor2Geo_FullName, Actor2Geo_CountryCode,  ActionGeo_FullName, ActionGeo_ADM1Code,
ActionGeo_ADM2Code, SOURCEURL, MentionTimeDate, MentionType, MentionSourceName, MentionIdentifier, Confidence, MentionDocLen, MentionDocTone
FROM events_and_eventmentions
WHERE ActionGeo_ADM1Code like 'WQ%' OR 
    ActionGeo_ADM1Code like 'CI%' OR
    ActionGeo_ADM1Code like 'PC%' OR
    ActionGeo_ADM1Code like 'NF%' OR
    ActionGeo_ADM1Code like 'CQ%' OR
    ActionGeo_ADM1Code like 'EC%' OR
    ActionGeo_ADM1Code like 'OS%' OR
    ActionGeo_ADM1Code like 'TL%' OR
    ActionGeo_ADM1Code like 'JQ%' OR
    ActionGeo_ADM1Code like 'HQ%' OR
    ActionGeo_ADM1Code like 'WF%' OR
    ActionGeo_ADM1Code like 'AS%' OR
    ActionGeo_ADM1Code like 'LQ%' OR
    ActionGeo_ADM1Code like 'FQ%' OR
    ActionGeo_ADM1Code like 'JA%' Or
    ActionGeo_ADM1Code like 'KQ%' OR
    ActionGeo_ADM1Code like 'DQ%' OR
    ActionGeo_ADM1Code like 'AQ%' OR
    ActionGeo_ADM1Code like 'GQ%' OR
    ActionGeo_ADM1Code like 'NC%' 
"""

# Execute the query and save the result
temp_result = engine.execute(query).fetchall()
# Convert result into numpy array
temp1 = np.array(temp_result)
# Convert np array into dataframe
territories = pd.DataFrame(data=temp1, columns=['GLOBALEVENTID', 'SQLDATE', 'MonthYear', 'Actor1Name', 'Actor1CountryCode', 'Actor2Name', 'Actor2CountryCode', 'EventCode', 'GoldsteinScale', 'NumMentions' ,'NumSources', 'NumArticles', 'AvgTone', 'Actor1Geo_FullName', 'Actor1Geo_CountryCode', 'Actor2Geo_FullName', 'Actor2Geo_CountryCode',  'ActionGeo_FullName', 'ActionGeo_ADM1Code', 'ActionGeo_ADM2Code', 'SOURCEURL', 'MentionTimeDate', 'MentionType', 'MentionSourceName', 'MentionIdentifier', 'Confidence', 'MentionDocLen', 'MentionDocTone'])
# Convert the datatypes from Object 
territories = territories.convert_dtypes()
# Show the top 10 results
territories.head(10)
```

### Who are the top URL sources of Oceania?

```python
query = """
SELECT MentionSourceName, count(*) as count
FROM events_and_eventmentions
GROUP BY MentionSourceName
ORDER BY count DESC
"""

# Execute the query and save the result
temp_result = engine.execute(query).fetchall()
# Convert result into numpy array
temp1 = np.array(temp_result)
# Convert np array into dataframe
df2 = pd.DataFrame(data=temp1, columns=['MentionSourceName', 'count'])
# Convert the datatypes from Object 
df2 = df2.convert_dtypes()
# Show the top 30 results
df2.head(30)
```

### Who are the top URL sources for Oceania Territories?

```python
engine = create_engine('sqlite://', echo=False)
territories.to_sql('territories', con=engine)

query = """
SELECT MentionSourceName, count(*) as count
FROM territories
GROUP BY MentionSourceName
ORDER BY count DESC
"""

# Execute the query and save the result
temp_result = engine.execute(query).fetchall()
# Convert result into numpy array
temp1 = np.array(temp_result)
# Convert np array into dataframe
df2 = pd.DataFrame(data=temp1, columns=['MentionSourceName', 'count'])
# Convert the datatypes from Object 
df2 = df2.convert_dtypes()
# Show the top 30 results
df2.head(30)
```

### Who are the top URL sources for Oceania Territories where the ActionGeo_ADM1Code is not Actor2Geo_CountryCode or Actor1Geo_CountryCode

```python
query = """
SELECT MentionSourceName, count(*) as count
FROM territories
WHERE ((SUBSTR(ActionGeo_ADM1Code, 1,2) IS NOT SUBSTR(Actor1Geo_CountryCode,1,2)) OR (SUBSTR(ActionGeo_ADM1Code, 1,2) IS NOT SUBSTR(Actor2Geo_CountryCode, 1,2)))
GROUP BY MentionSourceName
ORDER BY count DESC
"""

# Execute the query and save the result
temp_result = engine.execute(query).fetchall()
# Convert result into numpy array
temp1 = np.array(temp_result)
# Convert np array into dataframe
df2 = pd.DataFrame(data=temp1, columns=['MentionSourceName', 'count'])
# Convert the datatypes from Object 
df2 = df2.convert_dtypes()
# Show the top 30 results
df2.head(30)
```
